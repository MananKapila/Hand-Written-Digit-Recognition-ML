 <h2> STEP 4: Build the Model </h2>
<ol> 
<li> Set up the input layer, hidden layer and the output layer
<li> Compile the model
</ol>    

<p>The basic building block of neural network (in the <b>Keras</b> library) is the <b> layer </b>. Each layer consists of several pereptrons which are connected to the adjacent.</p>
<p> The configration of each layer depends on the data and the problem at hand </p>

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,activation='sigmoid'), #128 an arbitrary number
    keras.layers.Dense(10,activation='softmax')  
])
```

Structure of Weights

```python
hidden_layer= model.layers[1]
weights=hidden_layer.get_weights()
print('Shape of weights: ',np.shape(weights[0]))
print('Shape of baises: ',np.shape(weights[1]))
```

```python
output_layer= model.layers[2]
weights=output_layer.get_weights()
print('Shape of weights: ',np.shape(weights[0]))
print('Shape of baises: ',np.shape(weights[1]))
```

<h3> Compile the model </h3>
<p> Before training the model, few steps ae required:
    <ul>
        <li> Loss function: Checks the accuracy. We want to <b> minimize this function </b> to steer the model into right direction.We will use <b> Sparse Categorical Cross Entropy </b> </li>
        <li> Optimizer: This is how the model is updated based on the data . We will use <b> Stocastic Gradient Descent </b></li>
        <li> Metrics: Used to monitor training and testing steps. </li>
     </ul>
</p>            

```python
sgd=keras.optimizers.SGD(lr=0.5,decay=1e-6,momentum=0.5) #lr-> learining rate,it controls how much rate is chaged during each update, momentum is tech to drive sgd faster to optimal state. 
```

```python
model.compile(optimizer=sgd,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

<h3> Train the Model

```python
history=model.fit(train_images,train_labels,epochs=10,batch_size=100,validation_split=0.1) #epochs determines no. of iterations
#10% of training data is validation data
```

Above we can see that as the number of iterations increases the loss decreases and accuracy increases, the graph below also explains the same.

<h3> Visualize the model training

```python
val_losses=history.history['val_loss'] #history more like dictionary for fetching above training sample by name
losses=history.history['loss']
indicies=range(len(losses))

plt.figure(figsize=(10,5))
plt.plot(indicies,val_losses,color='r')
plt.plot(indicies,losses,color='g')
plt.legend(['Validation loss','Loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
```

Here we see an intersection point between the loss and validation loss, it tells that at certain point the loss of the training set is less than the loss of the validation data set in other words the accuracy of the training is more than than the accuracy of the validation, such a scenerio is called <b> overfitting </b>. The model has begun to memorise the training dataset, if training loss is much lower tha the validation loss, then we can say the extent of overfitting is high. When the validation loss is oly slightly higher tha the training loss, then is the ideal point to stop training the model.
<p> Note :) Overfitting happens when a model learns the detail and noise in the training data to the extent that it negatively impacts the performance of the model on new data. This means that the noise or random fluctuations in the training data is picked up and learned as concepts by the model. The problem is that these concepts do not apply to new data and negatively impact the models ability to generalize. </p>

<h3> Evaluate the model by computing the accuracy `

```python
test_loss,test_acc=model.evaluate(test_images,test_labels)
print('Test_accuracy:',test_acc) #97% accuracy
```

```python
predictions=model.predict(test_images)
```

<h3> Define a function to display image along with confidence levels

```python
def plot_confidence(images,labels,predictions):
    plt.figure(figsize=(15,30))
    plt.subplots_adjust(top=0.99,bottom=0.01,hspace=1.5,wspace=0.4)
    plot_index=0
    for i in range(len(images)):
        plot_index+=1
        plt.subplot(len(images),2,plot_index)
        plt.imshow(images[i],cmap=plt.cm.binary)
        correct_label=str(labels[i])
        predicted_label=str(np.argmax(predictions[i]))
        title='Correct Label: '+str(labels[i])+'\n'+'Predicted Label: '+str(np.argmax(predictions[i]))
        
        if predicted_label != correct_label:
            plt.title(title,backgroundcolor='r',color='w')
        else:
            plt.title(title,backgroundcolor='g',color='w')
        
        plt.xticks([])
        plt.yticks([])
        
        plot_index+=1
        plt.subplot(len(images),2,plot_index)
        plt.bar(range(10),predictions[i])
        plt.xticks(range(10))
        plt.ylim(0,1)
```

Plot 1st 10 images (correct labels)

```python
images=test_images[:10]
labels=test_labels[:10]
test_predictions=predictions[:10]
plot_confidence(images,labels,test_predictions)
```

Plot incorrect labels

```python
incorrect_indices=list()
for i in range(len(predictions)):
    predicted_label=np.argmax(predictions[i])
    if predicted_label!=test_labels[i]:
        incorrect_indices.append(i)
print('Number of incorrectly classified images: ',len(incorrect_indices)) 

incorrect_indices=incorrect_indices[:10]
incorrect_images= [test_images[i] for i in incorrect_indices]
incorrect_labels= [test_labels[i] for i in incorrect_indices]
incorrect_predictions=[predictions[i] for i in incorrect_indices]

plot_confidence(incorrect_images,incorrect_labels,incorrect_predictions)
```
